```
#MAX = x1*0.5 + x2*0.4
#Sujeito a:
#Açucar: x1+x2 <= 10000
#Amendoas: x1+x2 <= 2000
#Chocolate x1+x2 <= 3000
#Doce 1: 0.2x1 >= 2000
#Doce 2: 0.1x2 >= 2000
#        0.1x2 >= 3000 

# Definindo o problema como de maximização
prob = pulp.LpProblem('Exercício 1 da Lista 5', pulp.LpMaximize)

# Definindo as variáveis de decisão
x1 = pulp.LpVariable('Doce 1', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('Doce 2', lowBound=0, cat='Continuous')

#Definindo a função a ser minimizada
MAX = x1*0.5 + x2*0.4

#Adicionando a função-objetivo
prob += MAX

#Definindo a função do Açucar
acucar = x1+x2 

#Adicionando a função do Açucar nas restrições
prob += (acucar<=10000)

#Definindo a função das Amendoas
amendoas = x1+x2 

#Adicionando a função das Amendoas nas restrições
prob += (amendoas<=2000)

#Definindo a função do Chocolate
chocolate = x1+x2 

#Adicionando a função do Chocolate nas restrições
prob += (chocolate<=3000)

#Definindo a função do doce 1
doce1 = 0.2*x1 

#Adicionando a função do Açucar nas restrições
prob += (doce1>=2000)

#Definindo a função do doce 1
doce2 = 0.1*x2 

#Adicionando a função do Açucar nas restrições
prob += (doce2>=2000)

#Definindo a função do doce 1
doce2 = 0.1*x2 

#Adicionando a função do Açucar nas restrições
prob += (doce2>=3000)

#escrevendo o problema de otimização linear
print(prob)

# Resolvendo o problema 
optimization_result = prob.solve()

# Verificando se a solução ótima foi encontrada
#assert optimization_result == pulp.LpStatusOptimal

#mostrando o resultado
for var in (x1, x2):
    print('Produção semanal ótima de {}: {:1.0f}'.format(var.name, var.value()))
```

```
#Letra a do Exercício 2

#MIN = 20x1 + 20x2 + 31x3 + 11x4 + 12x5 
#Vitamina A = x1 + x3 + x4 + 2*x5 >= 15
#Vitamina K = x2 + 2*x3 + x4 + x5 >= 12

# Definindo o problema como de maximização
prob = pulp.LpProblem('Exercício 1 da Lista 5', pulp.LpMinimize)

# Definindo as variáveis de decisão
x1 = pulp.LpVariable('Nutriente 1', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('Nutriente 2', lowBound=0, cat='Continuous')
x3 = pulp.LpVariable('Nutriente 3', lowBound=0, cat='Continuous')
x4 = pulp.LpVariable('Nutriente 4', lowBound=0, cat='Continuous')
x5 = pulp.LpVariable('Nutriente 5', lowBound=0, cat='Continuous')

#Definindo a função a ser minimizada
MIN = 20*x1 + 20*x2 + 31*x3 + 11*x4 + 12*x5

#Adicionando a função-objetivo
prob += MIN

#Definindo a função da Vitamina A
VitA = x1 + x3 + x4 + 2*x5

#Adicionando a função da Vitamina A nas restrições
prob += (VitA>=15)

#Definindo a função da Vitamina K
VitK = x2 + 2*x3 + x4 + x5

#Adicionando a função da Vitamina K nas restrições
prob += (VitK>=12)

#escrevendo o problema de otimização linear
print(prob)

# Resolvendo o problema 
optimization_result = prob.solve()

# Verificando se a solução ótima foi encontrada
assert optimization_result == pulp.LpStatusOptimal

#mostrando o resultado
for var in (x1, x2, x3, x4, x5):
    print('Dieta mínima ótima do {}: {:1.0f}'.format(var.name, var.value()))
```
